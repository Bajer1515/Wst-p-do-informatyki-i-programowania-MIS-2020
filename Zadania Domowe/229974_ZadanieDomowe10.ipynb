```python
class Hodometr:
    def __init__(self, podstawy, wskazania):
        if type(podstawy) != list or type(wskazania) != list:
            raise TypeError('Podstawy i wskazania muszą być listami!')
        if len(podstawy) != len(wskazania):
            raise ValueError('Podstawy i wskazania muszą mieć równą długość!')
        for p in podstawy:
            if p <= 1:
                raise ValueError('Wszystkie podstawy muszą być większe równe 1!')
        for i in range(len(wskazania)):
            if wskazania[i] < 0 or wskazania[i] > podstawy[i]-1:
                raise ValueError('Nieprawidłowe wartości wskazań!')
        self.p = podstawy
        self.w = wskazania


    def __str__(self):
        m = []
        n = []
        for i in range(len(self.p)):
            if self.p[i] < 9:
                n.append(3)
            else:
                n.append(4)
            if self.w[i] < 9:
                m.append(3)
            else:
                m.append(4)    
        if len(self.p) == 1:
            w1 = '|{:^{}}|'.format(self.w[0],m[0])
            w2 = '|{:^{}}|'.format(self.p[0],n[0])
        elif len(self.p) == 2:
            w1 = '|{:^{}}|{:^{}}|'.format(self.w[1],m[1],self.w[0],m[0])
            w2 = '|{:^{}}|{:^{}}|'.format(self.p[1],n[1],self.p[0],[0])
        elif len(self.p) == 3:
            w1 = '|{:^{}}|{:^{}}|{:^{}}|'.format(self.w[2],m[2],self.w[1],m[1],self.w[0],m[0])
            w2 = '|{:^{}}|{:^{}}|{:^{}}|'.format(self.p[2],n[2],self.p[1],n[1],self.p[0],n[0])
        elif len(self.p) == 4:
            w1 = '|{:^{}}|{:^{}}|{:^{}}|{:^{}}|'.format(self.w[3],m[3],self.w[2],m[2],self.w[1],m[1],self.w[0],m[0])
            w2 = '|{:^{}}|{:^{}}|{:^{}}|{:^{}}|'.format(self.p[3],n[3],self.p[2],n[2],self.p[1],n[1],self.p[0],n[0])
        else:
            w1 = '|{:^{}}|{:^{}}|{:^{}}|{:^{}}|{:^{}}|'.format(self.w[-1],m[-1],self.w[-2],m[-2],'...',3,
                                                                self.w[1],m[1],self.w[0],m[0])
            w2 = '|{:^{}}|{:^{}}|{:^{}}|{:^{}}|{:^{}}|'.format(self.p[-1],n[-1],self.p[-2],n[-2],'...',3,
                                                                self.p[1],n[1], self.p[0],m[0])
        tabela = w1 + '\n' + w2
        return tabela

    
    # +
    def __add__(self,other):
        if isinstance(other, Hodometr):
            if self.p == other.p:
                z0 = (self.w[0]+other.w[0])%self.p[0]
                B = (self.w[0]+other.w[0])//self.p[0]
                # print('B0 = ',B)
                Z = [] + [z0]
                for i in range(1,len(self.p)):
                    z = (self.w[i]+other.w[i]+B)%self.p[i]
                    # print(f'{self.w[i]}+{other.w[i]}+{B} mod {self.p[i]} = {z}')
                    Z.append(z)
                    B = (self.w[i]+other.w[i]+B)//self.p[i]
                return Z
            else:
                raise ValueError('Podstawy obiektów muszą być takie same!')
        else:
            raise TypeError('Oba obiekty muszą być klasy Hodometr!')

    
    # -
    def __sub__(self,other):
        if isinstance(other, Hodometr):
            if self.p == other.p:
                z0 = (self.w[0]-other.w[0])%self.p[0]
                B = (self.w[0]-other.w[0])//self.p[0]
                Z = [] + [z0]
                for i in range(1,len(self.p)):
                    z = (self.w[i]-other.w[i]+B)%self.p[i]
                    Z.append(z)
                    B = (self.w[i]-other.w[i]+B)//self.p[i]
                return Z
            else:
                raise ValueError('Podstawy obiektów muszą być takie same!')
        else:
            raise TypeError('Oba obiekty muszą być klasy Hodometr!')


    # eq ==
    def __eq__(self,other):
        if isinstance(other, Hodometr):
            if self.p == other.p:
                wynik = []
                w1 = self.w[::-1]
                w2 = other.w[::-1]
                for i in range(len(self.p)):
                    wynik.append(w1[i] == w2[i])
                return wynik
                # if set(wynik)==True:
                #     return True
                # else:
                #     return False
            else:
                raise ValueError('Podstawy obiektów muszą być takie same!')
        else:
            raise TypeError('Oba obiekty muszą być klasy Hodometr!')
    

    # ne !=
    def __ne__(self,other):
        if isinstance(other, Hodometr):
            if self.p == other.p:
                wynik = []
                w1 = self.w[::-1]
                w2 = other.w[::-1]
                for i in range(len(self.p)):
                    wynik.append(w1[i] != w2[i])
                return wynik
            else:
                raise ValueError('Podstawy obiektów muszą być takie same!')
        else:
            raise TypeError('Oba obiekty muszą być klasy Hodometr!')


    # lt <
    def __lt__(self,other):
        if isinstance(other, Hodometr):
            if self.p == other.p:
                wynik = []
                w1 = self.w[::-1]
                w2 = other.w[::-1]
                for i in range(len(self.p)):
                    wynik.append(w1[i] < w2[i])
                return wynik
            else:
                raise ValueError('Podstawy obiektów muszą być takie same!')
        else:
            raise TypeError('Oba obiekty muszą być klasy Hodometr!')


    # le <=  
    def __le__(self,other):
        if isinstance(other, Hodometr):
            if self.p == other.p:
                wynik = []
                w1 = self.w[::-1]
                w2 = other.w[::-1]
                for i in range(len(self.p)):
                    wynik.append(w1[i] <= w2[i])
                return wynik
            else:
                raise ValueError('Podstawy obiektów muszą być takie same!')
        else:
            raise TypeError('Oba obiekty muszą być klasy Hodometr!')


    # gt >
    def __gt__(self,other):
        if isinstance(other, Hodometr):
            if self.p == other.p:
                wynik = []
                w1 = self.w[::-1]
                w2 = other.w[::-1]
                for i in range(len(self.p)):
                    wynik.append(w1[i] > w2[i])
                return wynik
            else:
                raise ValueError('Podstawy obiektów muszą być takie same!')
        else:
            raise TypeError('Oba obiekty muszą być klasy Hodometr!')

    
    # >=
    def __ge__(self,other):
        if isinstance(other, Hodometr):
            if self.p == other.p:
                wynik = []
                w1 = self.w[::-1]
                w2 = other.w[::-1]
                for i in range(len(self.p)):
                    wynik.append(w1[i] >= w2[i])
                return wynik
            else:
                raise ValueError('Podstawy obiektów muszą być takie same!')
        else:
            raise TypeError('Oba obiekty muszą być klasy Hodometr!')
```
